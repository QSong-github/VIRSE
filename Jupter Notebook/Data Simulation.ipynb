```python
import numpy as np
import pandas as pd

# Set random seed for reproducibility
np.random.seed(42)

# ==========================
# SIMULATION PARAMETERS
# ==========================
N = 500   # Number of samples (mutation reads)
D = 50    # Number of features (sequence length)
K = 3     # Number of clusters

# ==========================
# GENERATE GROUND TRUTH PARAMETERS
# ==========================
# Dirichlet distribution for cluster proportions
alpha = np.ones(K) * 2
pi_true = np.random.dirichlet(alpha)

# Beta distribution for mutation probabilities for each cluster
a, b = 2, 5
mu_true = np.random.beta(a, b, (K, D))

# Assign each sample to a cluster based on pi_true proportions
Z_true = np.random.choice(K, size=N, p=pi_true)

# Initialize mutation matrix X
X = np.zeros((N, D), dtype=int)

# Fill X based on mutation probabilities of each cluster
for n in range(N):
    cluster = Z_true[n]
    X[n, :] = np.random.binomial(1, mu_true[cluster, :])

# ==========================
# SAVE SIMULATED DATA & PARAMETERS
# ==========================
np.save("pi_true.npy", pi_true)
np.save("mu_true.npy", mu_true)
np.save("Z_true.npy", Z_true)

# Save the mutation matrix (X) and true clusters (Z_true) as a CSV
df = pd.DataFrame(X, columns=[f"Position_{i+1}" for i in range(D)])
df["Cluster"] = Z_true
df.to_csv("simulated_rna_mutations.csv", index=False)

# ==========================
# PRINT SUMMARY
# ==========================
print("✅ Simulation complete!")
print(f"- pi_true (cluster proportions): {pi_true}")
print(f"- mu_true shape (K x D): {mu_true.shape}")
print(f"- Z_true shape: {Z_true.shape}")
```
